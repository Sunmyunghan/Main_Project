```python
df_con = pd.concat([df_1, df_2], axis=0, ignore_index=True)
df_con
```

```python
df_con["dongname"] = 0
for i in range(len(df_con)):
    df_con["dongname"][i] = " ".join(df_con['address'].apply(lambda x:x.split(" ")).iloc[i][0:3])
df_con
```

```python
# 저장
df_con.to_csv("./속초-고성_맛집.csv", encoding='cp949', index=False)
```
